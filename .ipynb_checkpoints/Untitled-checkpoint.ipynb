```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
filename = 'sample.csv'
df = pd.read_csv(filename)
```

```python
df.head()
```

```python
df.info()
```

```python
data_graph = df['Month'].values
print(data_graph)
volume_graph = df['Volume'].values
print(volume_graph)
```

```python
plt.plot(data_graph, volume_graph)
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-darkgrid')
plt.plot(data_graph, volume_graph)
```

```python
plt.plot(data_graph, volume_graph, marker='^', markersize='5')
```

```python
plt.plot(data_graph, volume_graph, marker='*', markersize='8', color='b')
plt.xlabel('Month')
plt.ylabel('Volume')
```

```python
plt.plot(data_graph, volume_graph, marker = 'o', markersize = '7')
plt.xlabel('Month', fontsize=15)
plt.ylabel('Volume', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
```

```python
plt.plot(data_graph,volume_graph)
plt.grid(axis='y')
```

```python
plt.plot(data_graph,volume_graph, linestyle='-')
plt.grid(c='k', alpha = 0.4, linestyle='-')
```

```python
plt.plot(data_graph,volume_graph, linestyle='-', label = 'First')
plt.legend(loc=(1.02,0.5))
plt.grid(c='k', alpha = 0.4, linestyle='-')
```

```python
plt.plot(data_graph, volume_graph)
plt.xlabel('Month')
plt.ylabel('Volume')
plt.grid(color = 'g', alpha=0.5)
plt.savefig('graph.png', dpi=300)
```
